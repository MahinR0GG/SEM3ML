```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the provided CSV file
file_path = '/content/IRIS.csv'
iris_df = pd.read_csv(file_path)

# Create scatter plots
plt.figure(figsize=(12, 6))

# Scatter plot for sepal length and sepal width
plt.subplot(1, 2, 1)
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',marker='X', data=iris_df, hue='Species', palette='husl')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Scatter plot for petal length and petal width
plt.subplot(1, 2, 2)
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm',marker='X', data=iris_df, hue='Species', palette='husl')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('usedcars.csv')

fig, axs = plt.subplots(2,2, figsize=(10, 10))

# Histogram with distribution curve for mileage
sns.histplot(df['mileage'], kde=True, ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Mileage histogram with distribution Curve')
axs[0, 0].set_xlabel('Mileage')
axs[0, 0].set_ylabel('Frequency')

# Histogram with distribution curve for price
sns.histplot(df['price'], kde=True, ax=axs[0, 1], color='violet')
axs[0, 1].set_title('Price histogram with distribution Curve')
axs[0, 1].set_xlabel('Price')
axs[0, 1].set_ylabel('Frequency')

# Boxplot for mileage
sns.boxplot(x=df['mileage'], ax=axs[1, 0], color='blue')
axs[1, 0].set_title('Boxplot of Mileage')
axs[1, 0].set_xlabel('Mileage')

# Boxplot for price
sns.boxplot(x=df['price'], ax=axs[1, 1], color='violet')
axs[1, 1].set_title('Boxplot of Price')
axs[1, 1].set_xlabel('Price')

plt.show()
```
